## 1.01 Get Player Statistics
Individual player statistics acquired from [basketball-reference.com](basketball-reference.com) using the `basketball_reference_web_scraper` library. I researched and learned how to use this library from the documentation here: https://pypi.org/project/basketball-reference-web-scraper/

```python
import pandas as pd
import numpy as np
```

```python
from basketball_reference_web_scraper import client
```

```python
# initialize dictionary in which to append a dataframe of each year's player's statistics
dfs = {}
for i in range(1979,2020):
    dfs[i]=pd.DataFrame(client.players_season_totals(season_end_year=i))
```

```python
# add season column for each dataframe
for i in range(1979,2020):
    dfs[i]['season']=i-1
```

```python
for i in range(1979,2020):
    dfs[i]['team'] = dfs[i]['team'].astype(str)
```

```python
dfs[2019]
```

```python
teams = {}
for i in range(1979,2020):
    teams[i]=pd.DataFrame(dfs[i].groupby(['team']).sum())
    teams[i]['season']=i-1
```

```python
teams[1980]
```

```python
allyears = teams[1979]
for i in range(1980,2020):
    allyears = pd.concat([allyears,teams[i]])
```

```python
allyears
```

```python
allyears.to_csv('allyears_teams.csv')
```

```python
# create master dataset containing each individual season's dataset
allyears_players = dfs[1979]
for i in range(1980,2020):
    allyears_players = pd.concat([allyears_players,dfs[i]])
```

```python
allyears_players
```

```python
allyears_players.to_csv('allyears_players.csv')
```
