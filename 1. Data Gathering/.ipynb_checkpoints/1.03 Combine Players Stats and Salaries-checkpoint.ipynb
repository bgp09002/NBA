## 1.03 Combing Player 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv('allyears_players.csv')
salary = pd.read_csv('player_salary_data.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
stats.head()
```

```python
salary.head()
```

```python
salary.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
salary.dtypes
```

```python
salary['name_season']=salary['Player']+salary['Year'].astype(str)
```

```python
salary
```

```python
stats['name'] = stats['name'].map(lambda x: x.rstrip('*'))
```

```python
stats.head()
```

```python
stats['made_two_point_field_goals']=stats['made_field_goals']-stats['made_three_point_field_goals']
stats['attempted_two_point_field_goals']=stats['attempted_field_goals']-stats['attempted_three_point_field_goals']
stats['fg_pct']=stats['made_field_goals']/stats['attempted_field_goals']
stats['2fg_pct']=stats['made_two_point_field_goals']/stats['attempted_two_point_field_goals']
stats['3fg_pct']=stats['made_three_point_field_goals']/stats['attempted_three_point_field_goals']
```

```python
stats['ft_pct']=stats['made_free_throws']/stats['attempted_free_throws']
```

```python
stats['name_season']=stats['name']+stats['season'].astype(str)
```

```python
stats.drop(['Unnamed: 0', 'season1'], axis=1, inplace=True)
```

```python
stats
```

```python
stats.columns
```

```python
features = ['name', 'positions', 'age', 'team', 'games_played',
       'games_started', 'minutes_played', 'made_field_goals',
       'attempted_field_goals', 'made_three_point_field_goals',
       'attempted_three_point_field_goals', 'made_free_throws',
       'attempted_free_throws', 'offensive_rebounds', 'defensive_rebounds',
       'assists', 'steals', 'blocks', 'turnovers', 'personal_fouls', 'points',
       'season', 'made_two_point_field_goals',
       'attempted_two_point_field_goals', 'fg_pct', '2fg_pct', '3fg_pct',
       'name_season', 'ft_pct']
```

```python
stats1=stats.merge(salary, on='name_season', how='left')
```

```python
stats1
```

```python
stats1.drop(['slug', 'Player','Year'], axis=1, inplace=True)
```

```python
stats1 = stats1.dropna(axis=0, subset=['Salary'])
```

```python
stats1
```

```python
stats1.to_csv('players_stats_and_salaries.csv',index=False)
```
