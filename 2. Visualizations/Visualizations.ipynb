```python
# function to sort by desired stat and year

def sort_by_stat(stat,year,head=30):
    return teams_master[teams_master.index.year ==year][['TEAM_NAME',stat,'WINS','PLAYOFFS','FINALS_APPEARANCE','CHAMPS']].sort_values(by=[stat],ascending=False).head(head)

def sort_by_stat_multiyr(stat,year,head=30):
    return teams_master[teams_master.index.year >year][['TEAM_NAME',stat,'WINS','PLAYOFFS','FINALS_APPEARANCE','CHAMPS']].sort_values(by=[stat],ascending=False).head(head)
```

```python
players['salary'] = players['Salary'].str.replace(',', '')
players['salary'] = players['salary'].str.replace('$', '')
players['salary'] = players['salary'].astype(int)
```

```python
# creates dictionary of entire leagues salaries summed

league_tot_salary = players.groupby('season')['salary'].sum().to_dict()
```

```python
for year in range(1990,2019):
    players.loc[players.season == year, 'salary_normalized_%'] = players['salary']/league_tot_salary[year]*100
```

```python
players['year']=pd.to_datetime(players.season.astype(str))
```

```python
players.set_index('year', inplace=True)
```

```python
players['3PA/GP']=players['attempted_three_point_field_goals']/players['games_played']
```

```python
count_3pt_shooters=players.loc[(players['3fg_pct']>=.35) & (players['3PA/GP']>=3)].groupby(['season'])[['salary_normalized_%']].count()
```

```python
avg_sal_3pt_shooters=players.loc[(players['3fg_pct']>=.35) & (players['3PA/GP']>=3)].groupby(['season'])[['salary_normalized_%']].mean()
```
