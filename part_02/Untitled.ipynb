```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('all_teams_wins.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.tail(30)
```

```python
df['NBA_FINALS_APPEARANCE'].value_counts()
```

```python
df['FG2M']=df['FGM']-df['FG3M']
df['FG2A']=df['FGA']-df['FG3A']
df['FG2_PCT']=df['FG2M']/df['FG2A']
df['FG3A/FGA']=df['FG3A']/df['FGA']
df['FG2A/FGA']=df['FG2A']/df['FGA']
df['FTA/FGA']=df['FTA']/df['FGA']
df['FG3A/FG2A']=df['FG3A']/df['FG2A']
df['FG2M_PER_GP']=df['FG2M']/df['GP']
df['FG3M_PER_GP']=df['FG3M']/df['GP']
df['FG2A_PER_GP']=df['FG2A']/df['GP']
df['FG3A_PER_GP']=df['FG2A']/df['GP']
df['PTS_PER_GP']=df['PTS']/df['GP']
df['3PT_PER_PTS']=3*df['FG3M']/df['PTS']
df['2PT_PER_PTS']=2*df['FG2M']/df['PTS']
```

```python
df['PLAYOFFS'] = df.CONF_RANK.apply(lambda x: 1 if x < 9 else 0)
df['CHAMPS'] = df.NBA_FINALS_APPEARANCE.apply(lambda x: 1 if x == 'LEAGUE CHAMPION' else 0)
df['FINALS_LOSS'] = df.NBA_FINALS_APPEARANCE.apply(lambda x: 1 if x == 'FINALS APPEARANCE' else 0)
df['FINALS_APPEARANCE'] = df['CHAMPS']+df['FINALS_LOSS']
```

```python
df
```

```python
df['YEAR'] = df['YEAR'].str[:4]
```

```python
df['YEAR']=pd.to_datetime(df['YEAR'])
```

```python
league_scoring=df.groupby('YEAR').sum()[['FG2M','FG2A','FG3M','FG3A','FGM','FGA','FTM','FTA','PTS','GP']]
```

```python
league_scoring['FG2_PCT']=league_scoring['FG2M']/league_scoring['FG2A']
league_scoring['FG3_PCT']=league_scoring['FG3M']/league_scoring['FG3A']
league_scoring['FG_PCT']=league_scoring['FGM']/league_scoring['FGA']
league_scoring['FT_PCT']=league_scoring['FTM']/league_scoring['FTA']
league_scoring['FG3A/FGA']=league_scoring['FG3A']/league_scoring['FGA']
league_scoring['FG2A/FGA']=league_scoring['FG2A']/league_scoring['FGA']
league_scoring['FTA/FGA']=league_scoring['FTA']/league_scoring['FGA']
league_scoring['FG3A/FG2A']=league_scoring['FG3A']/league_scoring['FG2A']
league_scoring['FG2M_PER_GP']=league_scoring['FG2M']/league_scoring['GP']
league_scoring['FG3M_PER_GP']=league_scoring['FG3M']/league_scoring['GP']
league_scoring['FG2A_PER_GP']=league_scoring['FG2A']/league_scoring['GP']
league_scoring['FG3A_PER_GP']=league_scoring['FG3A']/league_scoring['GP']
league_scoring['PTS_PER_GP']=league_scoring['PTS']/league_scoring['GP']
league_scoring['3PT_PER_PTS']=3*league_scoring['FG3M']/league_scoring['PTS']
league_scoring['2PT_PER_PTS']=2*league_scoring['FG2M']/league_scoring['PTS']
```
